```python
#To display a pie chart of percentages of weather conditions (pie chart)
data.groupby['Weather Condition'].sum().plot(kind='pie', figsize=(5,5))
```

```python
#Q10: To display a pie chart of percentages of weather conditions
#pie chart
weather=data['Weather Condition'].value_counts()
dfweather=pd.DataFrame(weather)
dfweather = dfweather.reset_index()
dfweather.columns = ['Weather Condition', 'Frequency']
keep=dfweather[dfweather['Frequency']>20]
print(keep)
keep.plot.pie(y='Frequency',autopct='%1.1f%%', shadow=False, figsize=(12,9))
plt.title("Weather Conditions")
plt.show()
```
